```python
mean=round(np.mean(df,axis=1))
print(mean)
```

```python
cv =  round((np.std(df,axis=1) / np.mean(df,axis=1)),4)
print(cv)
```

```python
df.insert(15, "average", mean)
df.insert(16, "cv", cv)
df[:90]
```

```python
df.sort_values(by=['average'])
```
